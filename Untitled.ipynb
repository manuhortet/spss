```python
import pandas as pd
import pyreadstat

df, meta = pyreadstat.read_sav('PractiseDataFAKE.sav')
```

```python
column_names = [value for value in df.columns.values[1::] if value[3] == 'A']

clean_df = df[column_names].fillna(0)
correct_values_df = clean_df.round(0).astype(int).replace(8, 0).replace(3, 2)
correct_values_df['total_sum'] = correct_values_df.sum(axis=1)

df['total_sum'] = correct_values_df['total_sum']
df
```
